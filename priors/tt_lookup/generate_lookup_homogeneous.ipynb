```python
import os
if 'PATH_SET' not in locals():
    os.chdir('../..')
    PATH_SET = True
    
import torch    

if 'THREADS_SET' not in locals():
    try:
        os.environ['OMP_NUM_THREADS'] = '1'
        os.environ['MKL_NUM_THREADS'] = '1'
        
        torch.set_num_threads(1)
        torch.set_num_interop_threads(1)
        THREADS_SET = True
    except:
        pass

import numpy as np

import tqdm as tqdm

import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('https://raw.githubusercontent.com/dominik-strutz/dotfiles/main/mystyle_background.mplstyle')

figure_path = 'figures/priors/data_prior/'
os.makedirs(figure_path, exist_ok=True)
```

```python
from helpers.geographic_setup import (
    design_space_full, design_space_shoulder,
    events_shoulder, events_full
)

from helpers.helper_functions import (
    Homogeneous_Forward_Function
)
```

```python
prior_samples_full_disp = torch.load('data/priors/prior_samples_full_disp.pt')
prior_samples_full_uniform = torch.load('data/priors/prior_samples_full_uniform.pt')
prior_samples_shoulder_uniform = torch.load('data/priors/prior_samples_shoulder_uniform.pt')
```

```python
forward_function_homogeneous = Homogeneous_Forward_Function(
    v=1500.0
)
```

```python
full_disp_data = forward_function_homogeneous(
    prior_samples_full_disp, design_space_full)

full_uniform_data = forward_function_homogeneous(
    prior_samples_full_uniform, design_space_full)

shoulder_disp_data = forward_function_homogeneous(
    prior_samples_shoulder_uniform, design_space_shoulder)

shoulder_shots = forward_function_homogeneous(
    events_shoulder, design_space_shoulder)

lines_shots = forward_function_homogeneous(
    events_full, design_space_full)
```

```python
print(f'Memory size of full_disp_data: {full_disp_data.element_size() * full_disp_data.nelement() / 1024**2:.2f} MB')
print(f'Memory size of full_disp_data_uniform: {full_uniform_data.element_size() * full_uniform_data.nelement() / 1024**2:.2f} MB')
print(f'Memory size of shoulder_disp_data: {shoulder_disp_data.element_size() * shoulder_disp_data.nelement() / 1024**2:.2f} MB')
print(f'Memory size of shoulder_shots: {shoulder_shots.element_size() * shoulder_shots.nelement() / 1024**2:.2f} MB')
print(f'Memory size of lines_shots: {lines_shots.element_size() * lines_shots.nelement() / 1024**2:.2f} MB')
```
